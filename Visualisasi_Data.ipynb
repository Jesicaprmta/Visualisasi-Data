```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/covid19.csv')
df
```

```python
indonesia = df[df['Country/Region'] == 'Indonesia']
indonesia
```

```python
kasus = indonesia[indonesia['Confirmed']>0]
kasus
```

```python
kasusapril = kasus[kasus.Date >= '4/1/20']
kasusapril
```

```python
plt.plot(kasusapril.Date, kasusapril.Confirmed)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
plt.plot(kasusapril.Date, kasusapril.Confirmed)
plt.xticks(rotation = 'vertical')
plt.plot(kasusapril.Date, kasusapril.Recovered)
plt.show()
```

```python
plt.plot(kasusapril.Date, kasusapril.Confirmed)
plt.xticks(rotation = 'vertical')
plt.plot(kasusapril.Date, kasusapril.Recovered)
plt.plot(kasusapril.Date, kasusapril.Deaths)
plt.legend(['Jumlah kasus', 'Sembuh', 'meninggal'])
plt.show()
```
